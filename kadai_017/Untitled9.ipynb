```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self: object) -> bool:
    if self.age >= 20:
      print(f"{self.name}は大人の方です。")
    else:
      print(f"{self.name}は大人ではありません。")

Humans = [Human("森",20),Human("伊藤",21),Human("吉永",19),Human("青山",23)]

for human in Humans:
    human.check_adult()
```
